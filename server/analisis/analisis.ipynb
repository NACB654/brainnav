```python
import pandas as pd
import matplotlib.pyplot as plt

data_sonrinsa = pd.read_csv("../datos/sonrisa.csv")
data_sorpresa = pd.read_csv("../datos/sorpresa.csv")

print(data_sonrinsa.shape)
print(data_sorpresa.shape)
```

```python
sonrisa = data_sonrinsa[data_sonrinsa["lAct"] == "smile"]
sorpresa = data_sorpresa[data_sorpresa["uAct"] == "surprise"]
```

```python
sonrisa_mean = sonrisa["lPow"].mean()
sorpresa_mean = sorpresa["uPow"].mean()

print(sonrisa_mean, sorpresa_mean)
```

```python
plt.bar(["Sonrisa", "Levantamiento de cejas"], [sonrisa_mean, sorpresa_mean])

plt.title("Precision de detecci√≥n de gestos faciales")
plt.xlabel("Gestos")
plt.ylabel("Accuracy (%)")

plt.show()
```
