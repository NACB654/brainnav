```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("../datos/expresiones.csv")

data.shape
```

```python
sonrisa = data[data["lAct"] == "smile"]
sorpresa = data[data["uAct"] == "surprise"]

print(sonrisa.shape)
print(sorpresa.shape)
```

```python
sonrisa_mean = sonrisa["lPow"].mean()
sorpresa_mean = sorpresa["uPow"].mean()

print(sonrisa_mean, sorpresa_mean)
```

```python
plt.bar(["Sonrisa", "Levantamiento de cejas"], [sonrisa_mean, sorpresa_mean])

plt.title("Nivel de potencia de detecci√≥n de gestos faciales")
plt.xlabel("Gestos")
plt.ylabel("Accuracy (%)")

plt.show()
```
